2. Write a function that will plot the ranking of a given roller coaster over time as a line. Your function should take a roller coaster's name and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

```python
# 2
# Create a function to plot rankings over time for 1 roller coaster
def rcoaster_rankings(coaster_name,park_name, df):
    rankings_df = df[(df['name'] == coaster_name) & (df['park'] == park_name)] #2 cols are are stored in a single var., w/c contains the code to access specific values in the df.
    fig, ax = plt.subplots()
    ax.plot(rankings_df['year of rank'],rankings_df['rank'])
    ax.set_yticks(rankings_df['rank'].values)
    ax.set_xticks(rankings_df['year of rank'].values)
    ax.invert_yaxis()
    plt.title("{} Rankings".format(coaster_name))
    plt.xlabel('Year')
    plt.ylabel('Ranking')
    plt.show()
    plt.clf()
# Create a plot of El Toro ranking over time
rcoaster_rankings('El Toro','Six Flags Great Adventure', gtaw_woods)
```

##### Bar Chart for the Number of Inversions at an Amusement Park 

7. Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park. Your function should take the roller coaster DataFrame and an amusement park name as arguments. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame and amusement park name.

```python
# Create a function to plot inversions by coaster at park
# Create barplot of inversions by roller coasters
def coaster_inversions(park_name, coaster_df):
  park_coasters = coaster_df[(coaster_df['park']==park_name)].reset_index()
  park_coasters = park_coasters.sort_values('num_inversions', ascending = False)
  coaster_names = park_coasters['name']
  number_inversions = (park_coasters['num_inversions'])

  fig, ax = plt.subplots()
  plt.bar(range(len(number_inversions)), number_inversions) #should be range(len(x)) and there should be a height, otherwise error
  ax.set_xticks(range(len(coaster_names)))
  ax.set_xticklabels(coaster_names,rotation=90)
  #since the names are long, rotate to 90d for the txt to not overlap
  plt.title('Number of Inversions Per Coaster at {}'.format(park_name), fontsize = 14)
  plt.xlabel('Roller Coaster')
  plt.ylabel('# of Inversions')
  plt.show()
  plt.clf()

coaster_inversions('Parc Asterix', roller_coasters)
plt.clf()
```

###### Mean Speed, Length, and Height per Seating Type

```python
def seating_type_median(stype, df):
    stype_df = df[df.seating_type==stype]
    new_list = []
    mspeed = round(stype_df.speed.median(), 2)
    mlength = round(stype_df.length.median(), 2)
    mheight = round(stype_df.height.median(), 2)
    new_list.append([mspeed,mlength,mheight])
    return new_list

seating_type_median('Sit Down', roller_coasters)
```

```python
seating_type_median('Inverted', roller_coasters)
```

```python
seating_type_median('Spinning', roller_coasters)
```

##### Median Difference Between Two Numerical Variables Based on Seating Type

In this section, I will use the mean difference of the top two seating type, `Sit Down` and `Inverted`, to see if there is any association between a roller coaster's seating type and their speed, length, and height.

```python
def stype_median_diff(stype1,stype2, df):
    stype_df1 = df[df.seating_type==stype1]
    stype_df2 = df[df.seating_type==stype2]
    median_diff =[]
    mspeed = 'Median Difference in Speed: ' + str(round(stype_df1.speed.median(), 2) - round(stype_df2.speed.median(), 2))
    mlength = 'Median Difference in Length: ' + str(round(stype_df1.length.median(), 2) - round(stype_df2.length.median(), 2))
    mheight = 'Median Difference in Height: ' + str(round(stype_df1.height.median(), 2) - round(stype_df2.height.median(), 2))
    median_diff.append([mspeed, mlength, mheight])
    return median_diff

stype_median_diff('Sit Down', 'Inverted', roller_coasters)
```

###### Analysis
Based on the result above, the median difference of `Sit Down` and `Inverted` seating types is -5 for the speed, -151 for the length, and -9 for the height. It is important to note however than in the dataset, the number of roller coasters with inverted seating types are significantly lower compared to the number of roller coasters with a `Sit Down` seating type. Keep in mind that the extreme values also have an effect on the median of the speed, length, and height variables.

##### Scatterplot of Two Numerical Features based on Seating Type

In this section, I will analyze the relationship between a roller coaster's speed and height, as well as the relationship between their length and height, based on the roller coasters' seating type.

###### Speed vs. Height

```python
def shl_per_stype(var1, var2, df):
    df = df[(df['height'] < 140) & (df['length'] < 1750)] #I removed the extreme outliers to get a more accurate representation
    graph = sns.FacetGrid(df, col="seating_type", col_wrap=3,
                      hue = "seating_type", sharey = False, sharex = False)
    graph = (graph.map(sns.scatterplot,var1, var2).add_legend())
    
shl_per_stype('speed', 'height', roller_coasters)
```

```python
shl_per_stype('length', 'height', roller_coasters)
```

###### Analysis  

Based on the scatterplots above, it can be seen that the values for the `Sit down` type, the type with the highest number of observations, are clustering to the lower-left side of the graph. The value where these points cluster is similar to the `Inverted` type, which is within the range of __50-100 km/hr in speed, 500-1000 meter in length, 30-50 meter in height__.
Therefore it is safe to assume that different seating types do not result in a faster/higher/longer roller coasters.

### Log Transformation of Roller Coaster's Height

Since there are variables which are not normally distributed, I transformed those variables to make it more normally distributed. These might be useful if readers want to conduct linear regression, or other machine learning algorithms later on.Remember that a skewness of more than 1 means that the data is highly skewed.

```python
from sklearn.preprocessing import PowerTransformer
coaster_height = roller_coasters['height'].reset_index()
logtransform = PowerTransformer()
coaster_height['height'] = logtransform.fit_transform(coaster_height)
```

```python
plt.hist(coaster_height['height'], bins = 100)
plt.xlabel('Roller Coaster Height (in log)')
plt.ylabel('Number of Roller Coasters (in log)')
plt.title('Log Transformation of Roller Coasters\' Height')
plt.show()
plt.clf()
```
